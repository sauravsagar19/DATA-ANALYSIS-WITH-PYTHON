```python
## importing all the liberaries including pandas and numPy
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
```

```python
##opening the file
## and replacing "?" with NaN values
filename="imports-85.data"
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df = pd.read_csv(filename, names = headers)
df.replace("?",np.nan,inplace=True)
df.head()
```

```python
##checking for null values..isnull()=True where there is naN aur null values

missing_data = df.isnull()
missing_data.head(5)
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")  
```

```python
## Gives the average of normalized-losses

avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
##replacing NaN values of normalized-losses to the avg_norm_loss
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
##checking for replaced 
df.head()
## its all okkk :)
```

```python
avg_mean_bore=df["bore"].astype('float').mean(axis=0)
print('avg bore:',avg_mean_bore)
df["bore"].replace(np.nan,avg_mean_bore,inplace=True)  ## since bore had no NaN so it didnt replace anything
df.head()
```

```python
#Calculate the mean vaule for "stroke" column
avg_stroke = df["stroke"].astype("float").mean(axis = 0)
print("Average of stroke:", avg_stroke)

# replace NaN by mean value in "stroke" column
df["stroke"].replace("3.47", avg_stroke, inplace = True) ## tried to replace stroke having value
## 3.47 with the average value of stroke
df.head()
```

```python
avg_hp=df["horsepower"].astype("float").mean(axis=0)
print(avg_hp)
```

```python
df["horsepower"].replace(np.nan,avg_hp,inplace=True)
# df.head() ## for checking
```

```python
df['num-of-doors'].value_counts() ## to see the types of values in the table
```

```python
df["engine-location"].value_counts() 
```

```python
#replace the missing 'num-of-doors' values by the most frequent 
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
# simply drop whole row with NaN in "price" column
df.dropna(subset=["price"], axis=0, inplace=True)

# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
## Finally lets make the data pure and cleaned.....
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
df.dtypes ## finally checking
```

```python
df.head()
```

```python
## let us transform mpg to L/100km in the column of "highway-mpg", and change the name of column to "highway-L/100km"
df["highway-mpg"]=235/df["highway-mpg"]
df.rename(columns={"highway-mpg":"highway-L/100km"},inplace=True)
df.head()
```

```python
##According to the example above, normalize the column "height".
df["height"]=df["height"]/df["height"].max()
df["height"].head()
```

```python
## to draw histogram using matplotlib

df["horsepower"]=df["horsepower"].astype(int, copy=True)
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
group_names = ['Low', 'Medium', 'High']
```

```python
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
```

```python
df["horsepower-binned"].value_counts()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
##dummies example

dummy2=pd.get_dummies(df["aspiration"])
dummy2.rename(columns={"std":"updated_stud","turbo":"aspiration-turbo"},inplace=True)
dummy2.head()
```

```python
# merge the new dataframe to the original datafram
df = pd.concat([df, dummy2], axis=1)

# drop original column "aspiration" from "df"
df.drop('aspiration', axis = 1, inplace=True)
```

```python
df.to_csv('updated_df.csv')
```
